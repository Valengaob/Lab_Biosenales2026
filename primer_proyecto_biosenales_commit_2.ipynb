**1. CONSULTA**

**DescripciГіn del proyecto "EEG Motor Movement/Imagery":** Este proyecto se basa en una investigaciГіn de Interfaz Cerebro-Computador (BCI) que utiliza la ElectroencefalografГ­a (EEG) para capturar la forma en que el cerebro procesa el movimiento, es decir, se busca entender las seГ±ales elГ©ctricas del cerebro cuando una persona realiza tres tipos de tareas:

вЂў	Movimiento Real (EjecuciГіn motora): Cuando el sujeto mueve fГ­sicamente su mano izquierda o derecha.

вЂў	ImaginaciГіn de Movimiento (ImaginerГ­a motora): Cuando el sujeto solo piensa o imagina que estГЎ moviendo las manos, sin realizar ningГєn movimiento muscular.

вЂў	Reposo: cuando el sujeto esta tranquilo, sin pensar en movimientos (lГ­nea base)
El fin Гєltimo es verificar si un sistema computacional puede distinguir entre estos estados para, en el futuro, permitir que personas con movilidad reducida controlen prГіtesis o computadoras solo con el pensamiento.


**Historia del BCI:**

вЂўEl Origen (1924): Hans Berger descubre la electroencefalografГ­a (EEG) al registrar por primera vez la actividad elГ©ctrica del cerebro humano. IdentificГі las ondas Alpha (ritmo de Berger).

вЂўEl Nacimiento del tГ©rmino (1973): Jacques Vidal, de la UCLA, acuГ±a el tГ©rmino "Brain-Computer Interface". En su artГ­culo Toward Direct Brain-Computer Communication, planteГі que las seГ±ales de EEG podrГ­an usarse como un canal de control en tiempo real.

вЂўMГЎs adelante, entrados en la primera dГ©cada del 2000, Jonathan Wolpaw demostrГі que sujetos podГ­an mover un cursor en una pantalla usando ritmos sensoriomotores (Mu y Beta); mientras que, Phillip Kennedy implantГі el primer BCI intracortical en un paciente con sГ­ndrome de enclaustramiento, permitiГ©ndole controlar un cursor.

вЂўLa Era de la NeuroprГіtesis (2012): El proyecto BrainGate permitiГі que una mujer con tetraplejia controlara un brazo robГіtico para beber cafГ© usando un chip implantado en su corteza motora.

**Usos:**

вЂўComunicaciГіn: Para pacientes con ELA (Esclerosis Lateral AmiotrГіfica) o lesiones medulares, permitiГ©ndoles escribir en una pantalla (Spellers).

вЂўRestauraciГіn Motora: Control de exoesqueletos o prГіtesis biГіnicas mediante la detecciГіn de la imaginerГ­a motora.

вЂўNeuromarketing: Evaluar la respuesta emocional o atencional de un consumidor ante un producto sin que este tenga que hablar.

вЂўMonitoreo en tiempo real de la fatiga en pilotos o conductores para prevenir accidentes.

вЂўEntretenimiento: Videojuegos controlados por el estado de relajaciГіn o concentraciГіn del usuario (ej. cascos NeuroSky o Emotiv).

**Proyectos futuros en BCI:**

вЂўNeuralink (Elon Musk): Busca una alta densidad de electrodos (hilos ultra finos) y una cirugГ­a robГіtica para telepatГ­a digital y, a largo plazo, la simbiosis con la Inteligencia Artificial.

вЂўSincronizaciГіn Invasiva MГ­nima (Stentrode): Un dispositivo que se introduce por la vena yugular hasta el cerebro (sin abrir el crГЎneo) para registrar seГ±ales motoras. Ya estГЎ en ensayos clГ­nicos para pacientes con parГЎlisis.

вЂўBCI Bidireccional: No solo "leer" el cerebro, sino "escribir" en Г©l. Proyectos que buscan devolver el sentido del tacto a quienes usan prГіtesis enviando impulsos elГ©ctricos de vuelta a la corteza somatosensorial.

вЂўComputaciГіn Afectiva: Sistemas que ajusten el entorno (luz, mГєsica, dificultad de una tarea) detectando automГЎticamente el estado de ГЎnimo o el estrГ©s del usuario mediante EEG.

**DescripciГіn de la actividad cerebral registrada por EEG durante movimientos e imaginaciГіn de movimiento de las manos y cГіmo se hace su anГЎlisis en sistemas BCI:**

1. Ritmos Cerebrales de InterГ©s (Bandas de Frecuencia): En el anГЎlisis de BCI para tareas motoras, los ritmos mГЎs importantes cuya potencia (energia) se desea determinar se encuentran en la corteza sensoriomotora:

*   Ritmo Mu ($\mu$): Se localiza en el rango de 8 a 13 Hz. Es considerado el "ritmo de reposo" de la corteza motora. Cuando los mГєsculos estГЎn relajados, las neuronas disparan de manera sincronizada, generando una seГ±al de gran amplitud en esta frecuencia.

*   Ritmo Beta ($\beta$): Se encuentra entre los 13 y 30 Hz. Este ritmo estГЎ mГЎs vinculado a la planificaciГіn, ejecuciГіn de movimientos y al mantenimiento de la postura.



2. FenГіmenos de ModulaciГіn ERD y ERS:

*   ERD (DesincronizaciГіn Relacionada con Eventos): Cuando el sujeto empieza a mover o a imaginar el movimiento de una mano, las neuronas del ГЎrea motora correspondiente se desincronizan para procesar la informaciГіn. Esto resulta en una caГ­da drГЎstica de la potencia en las bandas Mu y Beta. Es, por definiciГіn, el marcador de "actividad".

*   ERS (SincronizaciГіn Relacionada con Eventos): Es el aumento de potencia que ocurre generalmente despuГ©s de finalizar el movimiento (rebote Beta) o cuando el sujeto entra en un estado de reposo absoluto. Indica que el sistema motor ha vuelto a su estado de "espera".

3. Electrodos y LocalizaciГіn (Sistema 10-20): Para captar la imaginerГ­a motora de las manos, el anГЎlisis se concentra en los electrodos situados sobre la Corteza Motora Primaria, siguiendo el sistema internacional 10-20:

*   C3 (Hemisferio Izquierdo): Controla y registra la actividad de la mano derecha (debido a la decusaciГіn de las vГ­as motoras).

*   C4 (Hemisferio Derecho): Controla y registra la actividad de la mano izquierda.

*   Cz (LГ­nea Media): Se asocia con el movimiento de las extremidades inferiores (pies) o integraciГіn motora general.

Ejemplo: Al analizar el dataset ds004362, si el sujeto imagina mover la mano derecha, se debe buscar un fenГіmeno de ERD (caГ­da de potencia) especГ­ficamente en el electrodo C3.

4. Flujo de AnГЎlisis en Sistemas BCI: El procesamiento de estas seГ±ales para convertirlas en una intenciГіn digital sigue estos pasos:

*   ReferenciaciГіn Espacial: Se suelen usar tГ©cnicas como el Laplaciano o CAR (Common Average Reference) para eliminar ruidos comunes a todos los electrodos y resaltar la actividad local de C3 y C4.

*   Filtrado Temporal: Se aplica un filtro pasa-banda (tГ­picamente entre 8 y 30 Hz) para aislar Гєnicamente los ritmos Mu y Beta, descartando parpadeos (baja frecuencia) y ruido muscular (alta frecuencia).

*   ExtracciГіn de CaracterГ­sticas (Features): PSD (Densidad Espectral de Potencia) y CSP (Common Spatial Patterns)

*   ClasificaciГіn: Se entrena un modelo (como LDA - AnГЎlisis Discriminante Lineal) para que aprenda a diferenciar los patrones de potencia y decida, en tiempo real, quГ© estГЎ intentando hacer el usuario.


**Flujos de procesamiento propuestos para estos tipos de proyectos y que se puedan realizar en el marco del curso:**

1. Pre-procesamiento (Limpieza de la seГ±al)
Las seГ±ales de EEG son extremadamente dГ©biles (ВµV) y estГЎn contaminadas por ruido ambiental y biolГіgico.

*   Filtrado Temporal: Se aplica un filtro pasa-banda (usualmente de 1 a 40 Hz), para eliminar la deriva de la lГ­nea base (0.5 Hz) y el ruido de alta frecuencia o actividad muscular (EMG).

*   Filtro Notch: EspecГ­ficamente en 60 Hz para eliminar la interferencia de la red elГ©ctrica.

*   Re-referenciaciГіn: Se suele usar el Promedio ComГєn (CAR) o una referencia Laplaciana sobre C3 y C4. Esto ayuda a resaltar la actividad local bajo el electrodo y eliminar el ruido que es comГєn a toda la cabeza.

2. SegmentaciГіn (Epoching)
Dado que el experimento se basa en tareas (imaginar mano derecha vs. izquierda), no analizamos la seГ±al continua, sino fragmentos de tiempo llamados Г‰pocas. Se deben extraer tambiГ©n Г©pocas de reposo para tener una lГ­nea base de comparaciГіn.

3. ExtracciГіn de CaracterГ­sticas (Feature Extraction)

*   PSD (Densidad Espectral de Potencia): Se calcula la energГ­a en las bandas Mu (8-13 Hz) y Beta (13-30 Hz).

*   AnГЎlisis Tiempo-Frecuencia (Espectrogramas): Permite ver cГіmo la potencia de estas bandas cae (ERD) justo cuando inicia la imaginaciГіn.

*   CSP (Common Spatial Patterns): (Opcional avanzado) Es un algoritmo que encuentra los mejores "filtros" para maximizar la diferencia de varianza entre la mano izquierda y derecha.

4. ClasificaciГіn o ComparaciГіn EstadГ­stica, se realiza una Inferencia EstadГ­stica:

*   CГЎlculo de Varianza/EnergГ­a: Se compara si la energГ­a en C3 durante la "imaginaciГіn derecha" es significativamente menor que durante el "reposo".

*   Pruebas de HipГіtesis: Uso de pruebas T-Student o Wilcoxon para validar si los cambios observados (ERD) no son producto del azar.

*   VisualizaciГіn de Topomapas: Crear mapas de calor de la cabeza para observar que la actividad se localiza efectivamente en la corteza motora.

**2. PLAN DE ANALISIS**

**Metodologia**

Se seguira la siguiente metodologia de procesamiento y analisis de la informaciГіn para evidenciar la diferencia en ritmos cerebrales asociados a las cuatro condiciones de estudio en los electrodos relevantes:

1. Pre-procesamiento y Limpieza (Acondicionamiento)Antes de buscar los ritmos, debemos asegurar que la seГ±al represente actividad cerebral pura y no ruido.

*   Filtrado Pasa-Banda: Aplicar un filtro entre 8 y 30 Hz para capturar ritmos Mu (8-13 Hz) y Beta (13-30 Hz), eliminando el ruido de baja frecuencia (parpadeos) y de alta frecuencia (EMG facial).

*   Filtro Notch (60 Hz): Para eliminar la interferencia de la red elГ©ctrica local.

*   Re-referenciaciГіn Espacial: Implementaremos un Promedio ComГєn (CAR) o Referencia Laplaciana centrada en C3, Cz y C4, con lo cual se busca eliminar el ruido global y resaltar la actividad local de la corteza motora.

2. SegmentaciГіn por Г‰pocas (Epoching): La seГ±al continua debe dividirse en fragmentos de tiempo basados en los eventos (triggers) del dataset:

DefiniciГіn de Clases:

*   Movimiento Real - Mano Izquierda.
*   Movimiento Real - Mano Derecha.
*   ImaginerГ­a Motora - Mano Izquierda.
*   ImaginerГ­a Motora - Mano Derecha.
*   LГ­nea Base (Reposo): Segmentos donde el sujeto no realiza tarea motora.

3. ExtracciГіn de CaracterГ­sticas (AnГЎlisis de Ritmos):

*   Densidad Espectral de Potencia (PSD): Utilizar el mГ©todo de Welch para calcular la potencia en las bandas Mu y Beta para cada electrodo (C3, Cz, C4).

*   CГЎlculo de ERD/ERS (DesincronizaciГіn/SincronizaciГіn): Calcular la potencia media durante la tarea ($P_{tarea}$), calcular la potencia media durante el reposo ($P_{reposo}$), aplicar la fГіrmula: $ERD\% = \frac{P_{tarea} - P_{reposo}}{P_{reposo}} \times 100$ e interpretar.

4. AnГЎlisis EstadГ­stico e Inferencia: el resultado de be mostrar una diferencia estadГ­sticamente significativa, lo cual se debera representar mediante.

*   Boxplots (Diagramas de caja) que comparen la potencia en C3 y C4 para las 4 condiciones contra el reposo.

*   Prueba de HipГіtesis: evaluaciГіn de normalidad  (Aplicar Shapiro-Wilk para decidir si usar pruebas paramГ©tricas) y comparaciones (con t-test (paramГ©trico) o Wilcoxon (no paramГ©trico)).

*   Topomapas: Crear mapas de calor de la cabeza para visualizar que la desincronizaciГіn ocurre realmente sobre el ГЎrea motora (contralateral al movimiento).

**Hipotesis**

HipГіtesis A: EjecuciГіn/ImaginaciГіn vs. Reposo (DetecciГіn de Actividad)

*   HipГіtesis Nula ($H_0$): No existe una diferencia significativa en la potencia de las bandas Mu/Beta entre el estado de reposo y la tarea (real o imaginada).

*   HipГіtesis Alternativa ($H_1$): La potencia en las bandas Mu/Beta disminuye significativamente durante la tarea debido a la DesincronizaciГіn Relacionada con Eventos (ERD).

HipГіtesis B: Lateralidad (Mano Derecha vs. Mano Izquierda)

*   HipГіtesis Nula ($H_0$): La disminuciГіn de potencia es simГ©trica en ambos hemisferios, independientemente de la mano utilizada.

*   HipГіtesis Alternativa ($H_1$): La disminuciГіn de potencia es mayor en el hemisferio contralateral a la mano utilizada (C3 para mano derecha, C4 para mano izquierda).

HipГіtesis C: Movimiento Real vs. ImaginaciГіn.

*   HipГіtesis Nula ($H_0$): Los niveles de ERD son idГ©nticos en el movimiento real y en la imaginaciГіn.

*   HipГіtesis Alternativa ($H_1$): El movimiento real genera un ERD mГЎs profundo y robusto que la simple imaginaciГіn del mismo.

**Herramientas:**

A. Procesamiento y ExtracciГіn de CaracterГ­sticas (Computacional)

*   **Numpy:** Para la manipulaciГіn de arreglos multidimensionales (Г©pocas, canales, muestras) sin usar ciclos for.

*   **Scipy (signal.welch):** Para calcular la PSD (Densidad Espectral de Potencia). Es la herramienta principal para cuantificar cuГЎnta energГ­a hay en los 8-30 Hz.

*   **MNE-Python:** Es la librerГ­a estГЎndar para EEG. Permite manejar archivos .fif o .set, filtrar y crear "Г©pocas" de manera eficiente.

B. AnГЎlisis EstadГ­stico (ValidaciГіn de HipГіtesis)Para decidir si rechazamos la $H_0$, usaremos:

*   **Prueba de Normalidad (Shapiro-Wilk):** Antes de elegir la prueba, debemos saber si tus datos de potencia siguen una distribuciГіn normal. Herramienta: scipy.stats.shapiro.

*   **Pruebas de ComparaciГіn:**

*   T-test de muestras relacionadas: Si los datos son normales.
*   Prueba de Wilcoxon: Si los datos no son normales (muy comГєn en EEG). Es la versiГіn no paramГ©trica del t-test.
Herramienta: scipy.stats.ttest_rel o scipy.stats.wilcoxon.

C. VisualizaciГіn (Evidencia GrГЎfica):Boxplots (Diagramas de Caja) y Topomapas (Topoplots)

Herramienta: mne.viz.plot_topomap.
