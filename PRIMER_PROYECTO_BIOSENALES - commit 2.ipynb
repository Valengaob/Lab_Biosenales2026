**1. CONSULTA**

**Descripción del proyecto "EEG Motor Movement/Imagery":** Este proyecto se basa en una investigación de Interfaz Cerebro-Computador (BCI) que utiliza la Electroencefalografía (EEG) para capturar la forma en que el cerebro procesa el movimiento, es decir, se busca entender las señales eléctricas del cerebro cuando una persona realiza tres tipos de tareas:

•	Movimiento Real (Ejecución motora): Cuando el sujeto mueve físicamente su mano izquierda o derecha.

•	Imaginación de Movimiento (Imaginería motora): Cuando el sujeto solo piensa o imagina que está moviendo las manos, sin realizar ningún movimiento muscular.

•	Reposo: cuando el sujeto esta tranquilo, sin pensar en movimientos (línea base)
El fin último es verificar si un sistema computacional puede distinguir entre estos estados para, en el futuro, permitir que personas con movilidad reducida controlen prótesis o computadoras solo con el pensamiento.


**Historia del BCI:**

•El Origen (1924): Hans Berger descubre la electroencefalografía (EEG) al registrar por primera vez la actividad eléctrica del cerebro humano. Identificó las ondas Alpha (ritmo de Berger).

•El Nacimiento del término (1973): Jacques Vidal, de la UCLA, acuña el término "Brain-Computer Interface". En su artículo Toward Direct Brain-Computer Communication, planteó que las señales de EEG podrían usarse como un canal de control en tiempo real.

•Más adelante, entrados en la primera década del 2000, Jonathan Wolpaw demostró que sujetos podían mover un cursor en una pantalla usando ritmos sensoriomotores (Mu y Beta); mientras que, Phillip Kennedy implantó el primer BCI intracortical en un paciente con síndrome de enclaustramiento, permitiéndole controlar un cursor.

•La Era de la Neuroprótesis (2012): El proyecto BrainGate permitió que una mujer con tetraplejia controlara un brazo robótico para beber café usando un chip implantado en su corteza motora.

**Usos:**

•Comunicación: Para pacientes con ELA (Esclerosis Lateral Amiotrófica) o lesiones medulares, permitiéndoles escribir en una pantalla (Spellers).

•Restauración Motora: Control de exoesqueletos o prótesis biónicas mediante la detección de la imaginería motora.

•Neuromarketing: Evaluar la respuesta emocional o atencional de un consumidor ante un producto sin que este tenga que hablar.

•Monitoreo en tiempo real de la fatiga en pilotos o conductores para prevenir accidentes.

•Entretenimiento: Videojuegos controlados por el estado de relajación o concentración del usuario (ej. cascos NeuroSky o Emotiv).

**Proyectos futuros en BCI:**

•Neuralink (Elon Musk): Busca una alta densidad de electrodos (hilos ultra finos) y una cirugía robótica para telepatía digital y, a largo plazo, la simbiosis con la Inteligencia Artificial.

•Sincronización Invasiva Mínima (Stentrode): Un dispositivo que se introduce por la vena yugular hasta el cerebro (sin abrir el cráneo) para registrar señales motoras. Ya está en ensayos clínicos para pacientes con parálisis.

•BCI Bidireccional: No solo "leer" el cerebro, sino "escribir" en él. Proyectos que buscan devolver el sentido del tacto a quienes usan prótesis enviando impulsos eléctricos de vuelta a la corteza somatosensorial.

•Computación Afectiva: Sistemas que ajusten el entorno (luz, música, dificultad de una tarea) detectando automáticamente el estado de ánimo o el estrés del usuario mediante EEG.

**Descripción de la actividad cerebral registrada por EEG durante movimientos e imaginación de movimiento de las manos y cómo se hace su análisis en sistemas BCI:**

1. Ritmos Cerebrales de Interés (Bandas de Frecuencia): En el análisis de BCI para tareas motoras, los ritmos más importantes cuya potencia (energia) se desea determinar se encuentran en la corteza sensoriomotora:

*   Ritmo Mu ($\mu$): Se localiza en el rango de 8 a 13 Hz. Es considerado el "ritmo de reposo" de la corteza motora. Cuando los músculos están relajados, las neuronas disparan de manera sincronizada, generando una señal de gran amplitud en esta frecuencia.

*   Ritmo Beta ($\beta$): Se encuentra entre los 13 y 30 Hz. Este ritmo está más vinculado a la planificación, ejecución de movimientos y al mantenimiento de la postura.



2. Fenómenos de Modulación ERD y ERS:

*   ERD (Desincronización Relacionada con Eventos): Cuando el sujeto empieza a mover o a imaginar el movimiento de una mano, las neuronas del área motora correspondiente se desincronizan para procesar la información. Esto resulta en una caída drástica de la potencia en las bandas Mu y Beta. Es, por definición, el marcador de "actividad".

*   ERS (Sincronización Relacionada con Eventos): Es el aumento de potencia que ocurre generalmente después de finalizar el movimiento (rebote Beta) o cuando el sujeto entra en un estado de reposo absoluto. Indica que el sistema motor ha vuelto a su estado de "espera".

3. Electrodos y Localización (Sistema 10-20): Para captar la imaginería motora de las manos, el análisis se concentra en los electrodos situados sobre la Corteza Motora Primaria, siguiendo el sistema internacional 10-20:

*   C3 (Hemisferio Izquierdo): Controla y registra la actividad de la mano derecha (debido a la decusación de las vías motoras).

*   C4 (Hemisferio Derecho): Controla y registra la actividad de la mano izquierda.

*   Cz (Línea Media): Se asocia con el movimiento de las extremidades inferiores (pies) o integración motora general.

Ejemplo: Al analizar el dataset ds004362, si el sujeto imagina mover la mano derecha, se debe buscar un fenómeno de ERD (caída de potencia) específicamente en el electrodo C3.

4. Flujo de Análisis en Sistemas BCI: El procesamiento de estas señales para convertirlas en una intención digital sigue estos pasos:

*   Referenciación Espacial: Se suelen usar técnicas como el Laplaciano o CAR (Common Average Reference) para eliminar ruidos comunes a todos los electrodos y resaltar la actividad local de C3 y C4.

*   Filtrado Temporal: Se aplica un filtro pasa-banda (típicamente entre 8 y 30 Hz) para aislar únicamente los ritmos Mu y Beta, descartando parpadeos (baja frecuencia) y ruido muscular (alta frecuencia).

*   Extracción de Características (Features): PSD (Densidad Espectral de Potencia) y CSP (Common Spatial Patterns)

*   Clasificación: Se entrena un modelo (como LDA - Análisis Discriminante Lineal) para que aprenda a diferenciar los patrones de potencia y decida, en tiempo real, qué está intentando hacer el usuario.


**Flujos de procesamiento propuestos para estos tipos de proyectos y que se puedan realizar en el marco del curso:**

1. Pre-procesamiento (Limpieza de la señal): según la literatura, lo que normalmente se hace en proyectos que involucran este tipo de dataset es hacer uso de la libreria mne para cargar los archivos del dataset, identificar los canales (electrodos) y eventos; sin embrago, las señales de EEG son extremadamente débiles (µV) y están contaminadas por ruido ambiental y biológico.

*   Filtrado Temporal: Se aplica un filtro pasa-banda (usualmente de 1 a 40 Hz, o de 8-30 Hz para analizar solo los ritmos mu y beta, vinculados al movimiento), para eliminar la deriva de la línea base (0.5 Hz) y el ruido de alta frecuencia (latidos del corazón) o actividad muscular (EMG, de los parpadeos por ejemplo).

*   Filtro Notch: Específicamente en 60 Hz para eliminar la interferencia de la red eléctrica.

*   Re-referenciación: Se suele usar el Promedio Común (CAR) o una referencia Laplaciana sobre C3 y C4. Esto ayuda a resaltar la actividad local bajo el electrodo y eliminar el ruido que es común a toda la cabeza.

2. Segmentación (Epoching)
Dado que el experimento se basa en tareas (imaginar mano derecha vs. izquierda), no analizamos la señal continua, sino fragmentos de tiempo llamados Épocas. Se deben extraer también épocas de reposo para tener una línea base de comparación.

3. Extracción de Características (Feature Extraction), siendo el punto en el que las señalas se pasan del dominio del tiempo al dominio de la frecuencia

*   PSD (Densidad Espectral de Potencia): Se calcula la energía en las bandas Mu (8-13 Hz) y Beta (13-30 Hz) y se determinan las frecuencias que dominan mediante el uso del metódo de Welch.

*   Análisis Tiempo-Frecuencia (Espectrogramas): Permite ver cómo la potencia de estas bandas cae (ERD) justo cuando inicia la imaginación. Por lo tanto, se busca analizar la desincronización relacionada con tales eventos en los electrodos motores (C3, C4 y Cz)

*   CSP (Common Spatial Patterns): Es un algoritmo que encuentra los mejores "filtros" para maximizar la diferencia de varianza entre la mano izquierda y derecha.

4. Clasificación o Comparación Estadística, se realiza una Inferencia Estadística:

*   Cálculo de Varianza/Energía: Se compara si la energía en C3 durante la "imaginación derecha" es significativamente menor que durante el "reposo".

*   Pruebas de Hipótesis: Uso de pruebas T-Student o Wilcoxon para validar si los cambios observados (ERD) no son producto del azar y comparar la potencia de señal entre "mano izquierda" y "reposo". Aquí se suelen emplear las pruebas de normalidad, como la Shapiro-Wilk, para decidir si usar estadística paramétrica o no.

*   Visualización de Topomapas: Crear mapas de calor de la cabeza para observar que la actividad se localiza efectivamente en la corteza motora.

5. Predicción de patrones: tipicamente se usan algoritmos como CSP (Common Spatial Patterns) o Redes Neuronales para predecir qué está pensando un nuevo sujeto. Este ya es un paso avanzado en cualquier proyecto de investigación empezando por los instrumentos de medición que implicaria tener.

**2. PLAN DE ANALISIS**

**Metodologia**

Se seguira la siguiente metodologia de procesamiento y analisis de la información para evidenciar la diferencia en ritmos cerebrales asociados a las cuatro condiciones de estudio en los electrodos relevantes:

1. Pre-procesamiento y Limpieza (Acondicionamiento)Antes de buscar los ritmos, debemos asegurar que la señal represente actividad cerebral pura y no ruido.

*   Filtrado Pasa-Banda: Aplicar un filtro entre 8 y 30 Hz para capturar ritmos Mu (8-13 Hz) y Beta (13-30 Hz), eliminando el ruido de baja frecuencia (parpadeos) y de alta frecuencia (EMG facial).

*   Filtro Notch (60 Hz): Para eliminar la interferencia de la red eléctrica local.

*   Re-referenciación Espacial: Implementaremos un Promedio Común (CAR) o Referencia Laplaciana centrada en C3, Cz y C4, con lo cual se busca eliminar el ruido global y resaltar la actividad local de la corteza motora.

2. Segmentación por Épocas (Epoching): La señal continua debe dividirse en fragmentos de tiempo basados en los eventos (triggers) del dataset:

Definición de Clases:

*   Movimiento Real - Mano Izquierda.
*   Movimiento Real - Mano Derecha.
*   Imaginería Motora - Mano Izquierda.
*   Imaginería Motora - Mano Derecha.
*   Línea Base (Reposo): Segmentos donde el sujeto no realiza tarea motora.

3. Extracción de Características (Análisis de Ritmos):

*   Densidad Espectral de Potencia (PSD): Utilizar el método de Welch para calcular la potencia en las bandas Mu y Beta para cada electrodo (C3, Cz, C4).

*   Cálculo de ERD/ERS (Desincronización/Sincronización): Calcular la potencia media durante la tarea ($P_{tarea}$), calcular la potencia media durante el reposo ($P_{reposo}$), aplicar la fórmula: $ERD\% = \frac{P_{tarea} - P_{reposo}}{P_{reposo}} \times 100$ e interpretar.

4. Análisis Estadístico e Inferencia: el resultado de be mostrar una diferencia estadísticamente significativa, lo cual se debera representar mediante.

*   Boxplots (Diagramas de caja) que comparen la potencia en C3 y C4 para las 4 condiciones contra el reposo.

*   Prueba de Hipótesis: evaluación de normalidad  (Aplicar Shapiro-Wilk para decidir si usar pruebas paramétricas) y comparaciones (con t-test (paramétrico) o Wilcoxon (no paramétrico)).

*   Topomapas: Crear mapas de calor de la cabeza para visualizar que la desincronización ocurre realmente sobre el área motora (contralateral al movimiento).

**Hipotesis**

Hipótesis A: Ejecución/Imaginación vs. Reposo (Detección de Actividad)

*   Hipótesis Nula ($H_0$): No existe una diferencia significativa en la potencia de las bandas Mu/Beta entre el estado de reposo y la tarea (real o imaginada).

*   Hipótesis Alternativa ($H_1$): La potencia en las bandas Mu/Beta disminuye significativamente durante la tarea debido a la Desincronización Relacionada con Eventos (ERD).

Hipótesis B: Lateralidad (Mano Derecha vs. Mano Izquierda)

*   Hipótesis Nula ($H_0$): La disminución de potencia es simétrica en ambos hemisferios, independientemente de la mano utilizada.

*   Hipótesis Alternativa ($H_1$): La disminución de potencia es mayor en el hemisferio contralateral a la mano utilizada (C3 para mano derecha, C4 para mano izquierda).

Hipótesis C: Movimiento Real vs. Imaginación.

*   Hipótesis Nula ($H_0$): Los niveles de ERD son idénticos en el movimiento real y en la imaginación.

*   Hipótesis Alternativa ($H_1$): El movimiento real genera un ERD más profundo y robusto que la simple imaginación del mismo.

**Herramientas:**

A. Procesamiento y Extracción de Características (Computacional)

*   **Numpy:** Para la manipulación de arreglos multidimensionales (épocas, canales, muestras) sin usar ciclos for.

*   **Scipy (signal.welch):** Para calcular la PSD (Densidad Espectral de Potencia). Es la herramienta principal para cuantificar cuánta energía hay en los 8-30 Hz.

*   **MNE-Python:** Es la librería estándar para EEG. Permite manejar archivos .fif o .set, filtrar y crear "épocas" de manera eficiente.

B. Análisis Estadístico (Validación de Hipótesis)Para decidir si rechazamos la $H_0$, usaremos:

*   **Prueba de Normalidad (Shapiro-Wilk):** Antes de elegir la prueba, debemos saber si tus datos de potencia siguen una distribución normal. Herramienta: scipy.stats.shapiro.

*   **Pruebas de Comparación:**

*   T-test de muestras relacionadas: Si los datos son normales.
*   Prueba de Wilcoxon: Si los datos no son normales (muy común en EEG). Es la versión no paramétrica del t-test.
Herramienta: scipy.stats.ttest_rel o scipy.stats.wilcoxon.

C. Visualización (Evidencia Gráfica): Boxplots (Diagramas de Caja) y Topomapas (Topoplots).

Herramienta: mne.viz.plot_topomap, matplotlib.pyplot y seaborn para los análisis estadísticos.
